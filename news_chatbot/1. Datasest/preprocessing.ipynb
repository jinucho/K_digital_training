```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
```

# 0. Dataset 전처리(Preprocessing)함수

```python
def generate_df(x: pd.DataFrame):
    # 리스트 컴프리헨션을 사용하여 모든 데이터를 먼저 리스트로 변환
    data = [
        [
            # doc['title'],
            # doc['id'],
            # [j['sentence'] for j in doc['text'][0]],
            # doc['abstractive'][0],
            # doc['extractive']
            (' ').join( [j['sentence'] for j in doc['text'][0]]),
            doc['abstractive'][0]
            
        ]
        for doc in x['documents']
    ]

    # 변환된 데이터를 사용하여 DataFrame 생성
    # temp = pd.DataFrame(data, columns=['media','id', 'article_original', 'abstractive', 'extractive'])
    temp = pd.DataFrame(data, columns=['news','summary'] )
    
    return temp
```
